L'istruzione **while**, utilizzata quando si desidera eseguire ripetutamente un blocco di codice finché una condizione specificata è vera. Continua l'iterazione finché la condizione non diventa falsa.

```python
# Questa funzione calcola il fattoriale di un numero naturale
# Input: un numero naturale
# Restituisce: il suo fattoriale
def fattoriale(n):
  result = 1
  while n > 0:
    result *= n
    n -= 1
  return result
```

```python
print(fattoriale(5))
```

L'istruzione **break** è un'istruzione di controllo del flusso utilizzata per uscire immediatamente dal ciclo in cui si trova. Il programma continua quindi l'esecuzione dalla prima istruzione dopo il ciclo.

```python
# Questa funzione determina se una chiave è contenuta in un dizionario
# Input: un dizionario e un valore
# Restituisce: (True, valore) se la chiave si trova nel dizionario, (False, "null") altrimenti
def cerca(dizionario, chiave):
  found=False
  valore="null"
  for element in dizionario:
    if element == chiave:
      found=True
      valore=dizionario[element]
      break
  if found:
    return (True, valore)
  else:
    return (False, valore)
```

```python
# Esempio
dizionario = {'a': 1, 'b': 2, 'c': 3}
print(cerca(dizionario, 'b'))
```

```python
print(cerca(dizionario, 'd'))
```

**Gestione delle eccezioni.** Le eccezioni possono essere gestite utilizzando i blocchi **try**, **except**, **else** e **finally**. Il blocco try contiene il codice che si vuole monitorare per le eccezioni.
Il blocco except viene eseguito quando si verifica un'eccezione all'interno del blocco try corrispondente. È possibile specificare quale tipo di eccezione catturare o semplicemente catturare tutte le eccezioni. È anche possibile catturare più eccezioni, fornendo una tupla di tipi di eccezione. Il blocco else viene eseguito se non si verificano eccezioni nel blocco try. Il blocco finally viene sempre eseguito, indipendentemente dal verificarsi di un'eccezione. Ecco un esempio:

```python
try:
  # Codice che può sollevare un'eccezione
  result = 10 / 0
except ZeroDivisionError:
  # Gestisce l'eccezione specifica (divisione per zero in questo caso)
  print("Errore: Divisione per zero!")
except Exception as e:
  # Gestisce qualsiasi altra eccezione
  print(f"Si è verificato un errore: {e}")
else:
  # Questo blocco viene eseguito se non si sono verificate eccezioni.
  print("Nessuna eccezione.")
finally:
  # Questo blocco viene sempre eseguito
  print("Blocco \"finally\" eseguito.")
```

La **comprensione di lista** (list comprehension) è un modo conciso per creare liste da iterabili, come liste, tuple o stringhe.

```python
# Creazione di una lista che eleva al quadrato gli elementi di un'altra lista
numbers = [1, 2, 3, 4, 5]
squared_numbers = [x ** 2 for x in numbers]
squared_numbers
```

```python
# Creazione di una lista che eleva al quadrato i numeri pari di un'altra lista
numbers = [1, 2, 3, 4, 5]
squared_even_numbers = [x ** 2 for x in numbers if x % 2 == 0]
squared_even_numbers
```

La **comprensione di dizionario** (dictionary comprehension) è simile alla comprensione di lista, ma produce un dizionario.

```python
# Creazione di un dizionario che associa a ogni elemento di una lista il suo quadrato
numbers = [1, 2, 3, 4, 5]
squared_dict = {x: x ** 2 for x in numbers}
squared_dict
```

```python
# Creazione di un dizionario che associa a ogni numero pari di una lista il suo quadrato
numbers = [1, 2, 3, 4, 5]
squared_even_dict = {x: x ** 2 for x in numbers if x % 2 == 0}
squared_even_dict
```

##Classi e oggetti
Python è un linguaggio di programmazione orientato agli oggetti

**Creare una classe.** Una classe è un tipo di oggetti. Ogni classe ha dei particolari *attributi* (variabili) che definiscono lo stato degli oggetti. Una classe ha anche funzioni, chiamate *metodi*, che accedono agli attributi e/o ne modificano i valori. I costruttori sono metodi speciali che creano un oggetto della classe. In Python, per creare una classe, si usa la parola chiave *class*. Statement def __init__(self, parameters) definisce un costruttore che crea un oggetto con i parametri dati.

```python
class Item ():
  '''
  Classe Item modella un oggetto trasportato in una supply chain
  '''
  def __init__(self, id, weight):
    '''
    Costruttore di un oggetto di classeItem
    :param id: un codice univoco associato all'oggetto
    :param weight: il peso dell'oggetto espresso in kg
    '''
    self.id = id
    self.weight = weight
  def __str__(self):
    return f"Oggetto {self.id} avente peso {self.weight} kg"
```

```python
a = Item('X23S', 1.5)
b = Item('X16A', 2.3)
print('Il peso dell\'item {} è {} kg'.format(a.id, a.weight))
print(b)
```

**Ereditarietà.** L'ereditarietà consente di definire una classe che eredita tutti i metodi e le proprietà da un'altra classe.
La classe genitore è la classe da cui viene ereditata, detta anche classe base. La classe figlio è la classe che eredita da un'altra classe, detta anche classe derivata. La funzione super() fa sì che la classe figlia erediti tutti i metodi e le proprietà dalla sua classe genitore.

```python
class Van():
  '''
  Questa clase modella un veicolo
  '''
  def __init__(self, plate_number, capacity, items):
    '''
    Costruttore di un oggetto di classe Van
    :param plate_number: il numero di targa del veicolo
    :param capacity: massimo peso che il veicolo può trasportare
    :param items: lista di oggetti di classe Item trasportati dal veicolo
    '''
    self.plate_number = plate_number
    self.capacity = capacity
    self.items = list(items)
    self.remaining_capacity = capacity
    for item in self.items:
      self.remaining_capacity -= item.weight

  def load(self, items):
    '''
    Questo metodo consente di caricare un elenco di articoli sul veicolo
    se il loro peso totale è inferiore o uguale alla capacità residua
    del veicolo. In caso affermativo, viene restituito True. In caso contrario,
    viene restituito False.
    :param items:
    :return:
    '''
    total_load = 0
    for item in items:
      total_load += item.weight
    if total_load > self.remaining_capacity:
      return False
    else:
      for item in items:
        self.items.append(item)
        self.remaining_capacity -= item.weight
      return True

  def __str__(self):
    stringa_items = ''
    for item in self.items:
      stringa_items += str(item) + ', '
    stringa_items = stringa_items[:-2]

    return f'''Veicolo targa {self.plate_number},
    capacità {self.capacity},
    capacità residua {self.remaining_capacity},
    oggetti presenti: {str(stringa_items)}'''
```

```python
class RefrigeratedVan (Van):
  '''
  Questa clase modella un veicolo con vano refrigerato. Eredita le caratteristiche
  della classe Van e aggiunge il metodo load_refrigerated, oltre agli attributi
  refrigerated_capacity, refrigerated_items e remaining_refrigerated_capacity.
  '''

  def __init__(self, plate_number, capacity, items, refrigerated_capacity, refrigerated_items):
    super().__init__(plate_number, capacity, items)
    self.refrigerated_capacity = refrigerated_capacity
    self.refrigerated_items = refrigerated_items
    self.remaining_refrigerated_capacity = refrigerated_capacity
    for item in self.refrigerated_items:
      self.remaining_refrigerated_capacity -= item.weight

  def load_refrigerated(self, items):
    '''
    Questo metodo consente di caricare un elenco di articoli sul veicolo nel vano
    refrigerato, se il loro peso totale è inferiore o uguale alla capacità residua
    del vano. In caso affermativo, viene restituito True. In caso contrario,
    viene restituito False.
    :param items:
    :return:
    '''
    total_load = 0
    for item in items:
      total_load += item.weight
    if total_load > self.remaining_refrigerated_capacity:
      return False
    else:
      for item in items:
        self.refrigerated_items.append(item)
        self.remaining_refrigerated_capacity -= item.weight
      return True

  def __str__(self):
    stringa_refrigerated_items = ''
    for item in self.refrigerated_items:
      stringa_refrigerated_items += str(item) + ', '
    stringa_refrigerated_items = stringa_refrigerated_items[:-2]

    stringa_items = ''
    for item in self.items:
      stringa_items += str(item) + ', '
    stringa_items = stringa_items[:-2]

    return f'''Veicolo refrigerato targa {self.plate_number},
    capacità {self.capacity},
    capacità residua {self.remaining_capacity},
    oggetti presenti: {stringa_items},
    capacità refrigerata {self.refrigerated_capacity},
    capacità residua refrigerata {self.remaining_refrigerated_capacity},
    oggetti refrigerati presenti: {stringa_refrigerated_items}'''
```

```python
van_1 = Van('AA000AA', 10, [])
print(str(van_1))

van_2 = RefrigeratedVan('BB111BB', 10, [a], 10, [b])
print(van_2)
```

```python
c = Item('X23L', 4.)
d = Item('X11W', 1.)
van_2.load([c])
van_2.load_refrigerated([d])

print(van_2)
```
